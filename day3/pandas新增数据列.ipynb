## pandas新增数据列

在数据分析中，很多时候现有的数据不能直接满足需求，需要新增数据列

新增数据列有4种方法
1. 直接赋值修改
2. apply方法修改
3. assigh方法修改
4. 分条件赋值

### 1.直接赋值修改

```python
import pandas as pd

df = pd.read_csv('new.csv')
df.set_index('ymd', inplace=True)
df.head()
```

```python
# 实例，增加一列，表示温差

df['wencha'] = df['high'] - df['low']
df.head()


#如上表达是 df['high'] 是一个series ，df['low'] 也是一个series， 相减法得到的也是一个series，最后赋值给 df['wencha']
```

```python
# 也可以用df.loc 实现修改列

df.loc[:, 'wencha'] = df['high'] - 2
df.head()
```
