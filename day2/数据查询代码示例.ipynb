## 4.使用条件表达式查询

bool列表的长度得等于行数或者列数

简单条件查询，最低温度低于-10度的列表

```python
result = df['low'] < -10

df.loc[result, :]
```

```python
# 取出所有最高温度大于30的所有天数
result = df['high'] > 30

df.loc[result, :]

```

```python
# 取出所有风向为 ‘西南风’ 的所有天数
result = df['fengxiang'] == '西南风'

df.loc[result, :]
```

```python
# 取出所有天气为 ‘阴~多云’ 的所有天数
result = df['tianqi'] == '阴~多云'

df.loc[result, :]
```

### 复杂天气查询，查一下2021年最完美的天气

***注意，组合条件用&符号合并，每个条件都得带括号***  & 

```python
## 查询最高温度小于30，最低温度大于15度， 并且天气为晴， 并且空气质量为为优  的天气

result = (df['high'] < 30) & (df['low'] > 15) & ((df['tianqi'] == '晴') | (df['tianqi'] == '多云~晴') ) & (df['aqiInfo'] == '优')

df.loc[result, :]
```

```python
## 查询最高温度小于35，最低温度大于20度， 并且天气为晴， 并且空气质量为为优或者轻度污染  的天气
result = (df['high'] < 35) & (df['low'] > 20) & (df['tianqi'] == '晴') & ((df['aqiInfo'] == '优') | (df['tianqi'] == '轻度污染'))

df.loc[result, :]
```
